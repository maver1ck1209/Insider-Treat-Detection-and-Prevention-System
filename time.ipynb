```python
logtime = [] 
for i in range(25):
    now = datetime.now()
    time = now.strftime("%H:%M:%S")
    date = now.strftime("%m/%d/%Y")
    t = time.split(":") + date.split("/")
    t[0] = '8'
    t = [int(x) for x in t]
    t
    a = random.randint(0,60)
    b = random.randint(0,60)
    t[1] += a
    t[2] += b
    if t[1]>=60:
        t[1]-=60
    if t[2]>=60:
        t[2]-=60
    t[4]+=i
    logtime.append(t)
logtime
```
