```python
from tkinter import *
from tkinter import messagebox
from PIL import ImageTk, Image
from datetime import date
from datetime import datetime
import joblib
import pickle
import pandas as pd
import cv2
from simple_facerec import SimpleFacerec
```

```python
facerec = SimpleFacerec()
facerec.load_encoding_images("images/")
```

```python
model = joblib.load('aics_KN.pkl')
```

```python
def extrapass(a):
    
    if a!=1:
        m=0
        root=Tk()
        root.title("Untimely Login")
        root.geometry("300x200")
        def login():
            username = entry1.get()
            password = entry2.get()
            if username=='aics' and password=='jcomp':
                messagebox.showinfo("", "Initiating Face Recognition")
                cap = cv2.VideoCapture(0)
                while True:
                    ret, frame = cap.read()
                    faceloc,facename = facerec.detect_known_faces(frame)
                    for i,j in zip(faceloc, facename):

                        y1,x2,y2,x1 = i[0],i[1],i[2],i[3] 
                        frame = cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),2)
                        frame = cv2.rectangle(frame,(x1,y2-35),(x2,y2),(0,255,0),cv2.FILLED)
                        frame = cv2.putText(frame,j,(x1+6,y2-6),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
                    cv2.imshow("WEBCAM", frame)
                    key = cv2.waitKey(1)
                    try:
                        if facename[0] == 'Giridhar':
                            m=0
                            messagebox.showinfo("", "Login Successful")
                            break
                    except:
                        pass
                    if key ==27:
                        m=1
                        messagebox.showinfo("", "Access Denied: Insider Threat Detected")
                        break
                cap.release()
                cv2.destroyAllWindows()

            elif username=='' and password=='':
                messagebox.showinfo("", "Empty Not allowed")
            else:
                messagebox.showinfo("", "Login Denied Insider Threat Detected")
        global entry1
        global entry2

        Label(root,text='Username').place(x=20,y=20)
        Label(root,text='Password').place(x=20,y=70)

        entry1 = Entry(root,bd=5)
        entry1.place(x=140,y=20)
        entry2 = Entry(root,bd=5)
        entry2.place(x=140,y=70)

        Button(root, text="Login", command = login, height=3, width=13, bd = 6).place(x=100,y=120)
        root.mainloop()
        return m
```
